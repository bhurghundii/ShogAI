```python
import matplotlib.pyplot as plt
import numpy as np

f = open("autoacc.txt", "r")
x = f.readlines()

fig, ax = plt.subplots()

plt.plot(x)

every_nth = 20
for n, label in enumerate(ax.yaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)

every_nth = 20
for n, tick in enumerate(ax.yaxis.get_ticklines()):
    if n % every_nth != 0:
        tick.set_visible(False)
        
plt.ylabel('Accuracy')
plt.xlabel('Epoch * Iterations')

plt.show()
```
